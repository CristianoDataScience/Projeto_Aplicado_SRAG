## Base de Dados

### Link da base de dados: <https://www.saopaulo.sp.gov.br/planosp/simi/dados-abertos/>

```python
df = pd.read_csv('/content/drive/MyDrive/Projeto_Aplicado_SRAG/srag.csv', sep=';')
df
```

```python
#Analisando tipo das colunas do DataFrame
df.info()
```

```python
#drop na coluna UF e Raça
df.drop(['SG_UF', 'CS_RACA'],axis=1, inplace=True)
df
```

```python
df
```

```python
#Renomeando as colunas do DataFrame
df = df.rename(columns={'DATA_DE_NOTIFICACAO': 'data', 'Municipio': 'minicipio', 'Faixa_Etaria': 'idade',
                              'CS_SEXO': 'sexo', 'Outras_SRAG': 'srag', 'EVOLUCAO': 'evolucao', 'Qtde_Masculino': 'qtde_masculino',
                              'Qtde_Feminino': 'qtde_feminino', 'Qtde_Total': 'total'},)
```

```python
df.head()
```

```python
# alterando tipo de coluna
df = df.astype({
    'data': 'datetime64'

})
df.info()
```

```python
#Analisando dados faltantes
df.isnull().sum()
```

```python
#Salvando DataFrame
df.to_csv('base_dados_srag_sp.csv', encoding = 'iso-8859-1', index= False)
```
