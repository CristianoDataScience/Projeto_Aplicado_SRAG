## Base de dados 

```python
df_sp = pd.read_csv('/content/drive/MyDrive/Projeto_Aplicado_SRAG/base_dados_srag_sp.csv',
                    sep=',', encoding='iso-8859-1')
# encoding: codificação de caracteres, normalmente utiliza-se o iso-8859-1, utf-8, latin-1)
df_sp.head()
```

```python
#renomendo as linhas 
df_sp['sexo'] = df_sp['sexo'].replace({'Masculino': 'M', 'Feminio': 'F'})
df_sp
```

## Base de Dados apenas Síndrome Respiratória Aguda Grave (SRAG)

```python
df_sp = df_sp[df_sp['srag'] != 'Covid-19']
df_sp = df_sp.rename(columns={'minicipio': 'municipio', 'idade': 'faixa_etaria'})
df_sp
```

```python
#Analisando quantidade de casos por idade 
df_sp['faixa_etaria'].value_counts().sort_index()
```

```python
#Análise estatística
df_sp.describe()
```

```python
df_sp['total'].value_counts().sort_index()
```

## Analisando casos de 0 há 15 anos por municipio

```python
menores_de_15 = df_sp.query('faixa_etaria <= 15')['municipio'].value_counts()
```

```python
menores_de_15
```

```python
df_sp_maiores = df_sp.loc[df_sp.faixa_etaria > 15]
df_sp_maiores
```
